```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

## Testing Classification

```python
dataset = pd.read_csv('data/classification/heart.csv')
dataset.head()
```

```python
X = dataset.drop(columns='target', axis=1)
Y = dataset['target']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# pip install ubml
from ubml.train_test import classification_train_test
classification_train_test(x_train=X_train, y_train=Y_train, x_test=X_test, y_test=Y_test)
```

## Testing Regression

```python
dataset = pd.read_csv('data/regression/car_dataset.csv')
dataset.head()
```

```python
X = dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y = dataset['Selling_Price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
# pip install ubml
from ubml.train_test import regression_train_test
regression_train_test(x_train=X_train, y_train=Y_train, x_test=X_test, y_test=Y_test)
```
